```python
import numpy as np
import pandas as pd
lali=pd.read_csv('Laliga.csv',index_col=0)
df=lali.replace('-', 0)
df.columns=df.iloc[0]
df.drop(df.index[0], inplace=True)
print(df)
```

```python
df['Debut'] = df['Debut'].astype(str).str[:4].astype(int)
debut=df[df['Debut'].between(1930, 1980)]
print(debut)
```

```python
df['Points'] = df['Points'].astype(str).str[:4].astype(int)
print(df.sort_values('Points',ascending=False).head(5))
```

```python
df['GoalsFor'] = df['GoalsFor'].astype(str).str[:4].astype(int)
df['GoalsAgainst'] = df['GoalsAgainst'].astype(str).str[:4].astype(int)
def Goal_diff_count(x,y):
    a=x-y
    df['Goal_diff_count']=a
    return Goal_diff_count
Goal_diff_count(df['GoalsFor'],df['GoalsAgainst'])
print(df[df.Goal_diff_count == df.Goal_diff_count.max()]) 
print(df[df.Goal_diff_count == df.Goal_diff_count.min()])
df=df.drop('Goal_diff_count',1)
```

```python
df['GamesWon'] = df['GamesWon'].astype(str).str[:4].astype(int)
df['GamesPlayed'] = df['GamesPlayed'].astype(str).str[:4].astype(int)
df['Winning_Percent']=(df['GamesWon']/df['GamesPlayed'])*100
df=df.fillna(0)
print(df.sort_values('Winning_Percent',ascending=False).head(5))
```

```python
df['BestPosition'] = df['BestPosition'].astype(str).str[:2].astype(int)
print(df.groupby(df['BestPosition'])['Points'].sum())
```
